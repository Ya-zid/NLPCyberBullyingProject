```python
import pandas as pd
from langdetect import detect

# Function to generate new variations by swapping characters
def generate_variations(word):
    variations = [word]

    # Detect the language of the word
    try:
        lang = detect(word)
    except:
        lang = 'unknown'

    # Arabic-specific swaps
    if lang == 'ar':
        swaps = {
          'ا': 'آ',  # Alef to Alif with a long sound
          'و': 'ؤ',  # Waw to Waw with Hamza
          'ع': 'غ',  # Ayn to Ghayn
          'ي': 'ئ',  # Yeh to Yeh with Hamza
          'ا': 'ى',  # Alif to Alef Maksura
          'و': 'ى',  # Waw to Alef Maksura (in some dialects)
          'ي': 'ى',  # Yeh to Alef Maksura (in some dialects)
          'ه': 'ة',  # Ha to Taa Marbuta
          'ق': 'ك',  # Qaf to Kaf (in dialects)
          'غ': 'خ',  # Ghayn to Khaa (in dialects)
          'ص': 'ض',  # Sinn to Sad (in dialects)
          'د': 'ز',  # Daal to Zay (in some informal cases)
          'ش': 'س',  # Shin to Seen (in some dialects)
          'ء': 'ا',  # Hamza to Alif (informally)
          'ث': 'ت',
        }
    # Latin (English) specific swaps
    elif lang == 'en':
        swaps = {
          'a': 'e',  # 'a' to 'e' in informal writing
          'e': 'a',  # 'e' to 'a' in informal writing
          'o': 'u',  # 'o' to 'u' in informal writing
          'u': 'o',  # 'u' to 'o' in informal writing
          'i': 'y',  # 'i' to 'y' in informal writing
          'i': 'ee',  # 'i' to 'ee' in informal writing
          'y': 'i',  # 'y' to 'i' in informal writing
          'y': 'ee',  # 'y' to 'ee' in informal writing
          'ou': 'o',  # 'ou' to 'o' in informal writing
          'o': 'ou',  # 'o' to 'ou' in informal writing
          'ou': 'oo',  # 'ou' to 'oo' in informal writing
          'ch': 'sh',  # 'ch' to 'sh' (common for Arabic 'ش')
          's': 'z',  # 's' to 'z' (common in Egyptian Arabic and other dialects)
          'kh': 'k',  # 'kh' to 'k' (common in some dialects)
          'q': 'k',  # 'q' to 'k' (common in some dialects)
          '3': 'ع',  # '3' for 'ع' (common in Arabizi, transliteration for 'Ayn')
          '7': 'ح',  # '7' for 'ح' (common in Arabizi, transliteration for 'Haa')
          '5': 'خ',  # '5' for 'خ' (common in Arabizi, transliteration for 'Khaa')
          '6': 'ط',  # '6' for 'ط' (common in Arabizi, transliteration for 'Taa')
          '9': 'ص',  # '9' for 'ص' (common in Arabizi, transliteration for 'Sadd')
          '8': 'غ',  # '8' for 'غ' (common in Arabizi, transliteration for 'Ghayn')
          '4': 'ف',  # '4' for 'ف' (common in Arabizi, transliteration for 'Fa')
          '2': 'ا',  # '2' for 'ا' (common in Arabizi, transliteration for 'Alif')
          '0': 'و',  # '0' for 'و' (common in Arabizi, transliteration for 'Waw')
        }
    else:
        swaps = {}

    for char, swap_char in swaps.items():
        if char in word:
            new_word = word.replace(char, swap_char)
            if new_word != word:
                variations.append(new_word)

    return variations

def generate_permuted_rows(df):
    new_rows = []

    for _, row in df.iterrows():
        word = row['text']
        variations = generate_variations(word)

        for variation in variations:
            new_row = {'text': variation, 'label': row['label']}
            new_rows.append(new_row)

    permuted_df = pd.DataFrame(new_rows)
    return pd.concat([df, permuted_df], ignore_index=True)


text = generate_permuted_rows(text)

text
```

```python
import pandas as pd
import re

def normalize_text(text):
    """
    Normalizes text containing Arabic and Latin scripts for NLP tasks,
    including the removal of redundant letters.

    Args:
        text (str): The input text to be normalized.

    Returns:
        str: The normalized text.
    """
    if not isinstance(text, str):
        return ""

    def normalize_arabic(text):
        text = re.sub("[إأآا]", "ا", text)
        text = re.sub("ى", "ي", text)
        text = re.sub("ة", "ه", text)
        text = re.sub("ؤ", "و", text)
        text = re.sub("ئ", "ي", text)
        text = re.sub("[ًٌٍَُِّْ]", "", text)
        text = re.sub("[ﻷﻵﻹﻳﻯ]", "ا", text)
        return text

    text = text.lower()

    text = re.sub(r"http\S+|www\S+", "", text)
    text = re.sub(r"@\w+", "", text)
    text = re.sub(r"#\w+", "", text)
    text = re.sub(r"[^\w\s\u0600-\u06FF\U0001F600-\U0001F64F]", "", text)
    text = re.sub(r'(.)\1{2,}', r'\1', text)

    text = re.sub(r"\s+", " ", text).strip()
    text = normalize_arabic(text)

    return text


data = pd.read_csv('final_data_with_aug.csv')

text_column = 'text'
data[text_column] = data[text_column].apply(normalize_text)
data.to_csv('normalized_data.csv', index=False)

print("Normalization complete. File saved as 'normalized_data.csv'.")
```

# **Fine Tuning :**





```python
import pandas as pd
import torch
from torch.utils.data import Dataset, DataLoader
from transformers import AutoTokenizer, AutoModelForSequenceClassification
from sklearn.model_selection import train_test_split
import torch
from transformers import AdamW
from tqdm import tqdm
from torch import nn
from transformers import AutoModel
from transformers import AutoTokenizer
from sklearn.metrics import classification_report
from peft import get_peft_model, LoraConfig, TaskType
```

```python
model_checkpoint = 'sentence-transformers/all-MiniLM-L6-v2'
tokenizer = AutoTokenizer.from_pretrained(model_checkpoint)
base_model = AutoModel.from_pretrained(model_checkpoint)
```

```python
# mapping the labels from words to integers
def label_mapper(row):
    if row['label'] == 'cyberbullying':
        return 1
    elif row['label'] == 'not cyberbullying':
        return 0
    elif row['label'] == 'context based':
        return 2
    else:
        raise ValueError(f"Unexpected label: {row['label']}")

df['label'] = df.apply(label_mapper, axis=1)


# asserting the types of the columns
df["label"] = df["label"].astype(int)
df["text"] = df["text"].fillna("").astype(str)
df["bad_words_count"] = df["bad_words_count"].fillna(0).astype(float)
```

```python
# splitting the dataset to train set, validation set and test set
train_df, temp_df = train_test_split(df, test_size=0.2, random_state=42, stratify=df["label"])
val_df, test_df = train_test_split(temp_df, test_size=0.5, random_state=42, stratify=temp_df["label"])
```

```python
max_length = 128

# dataset wrapper
class CyberbullyingDataset(Dataset):
    def __init__(self, texts, labels, bad_words_count, tokenizer, max_length):
        self.texts = texts
        self.labels = labels
        self.bad_words_count = bad_words_count
        self.tokenizer = tokenizer
        self.max_length = max_length

    def __len__(self):
        return len(self.texts)

    def __getitem__(self, idx):
        text = self.texts[idx]
        label = self.labels[idx]
        bad_words = self.bad_words_count[idx]

        #tokenization of inputs
        encoding = self.tokenizer(
            text,
            max_length=self.max_length,
            padding="max_length",
            truncation=True,
            return_tensors="pt"
        )

        return {
            "input_ids": encoding["input_ids"].squeeze(0),
            "attention_mask": encoding["attention_mask"].squeeze(0),
            "label": torch.tensor(label, dtype=torch.long),
            "bad_words_count": torch.tensor(bad_words, dtype=torch.float),
        }
```

```python
# datasets creation
train_dataset = CyberbullyingDataset(
    texts=train_df["text"].tolist(),
    labels=train_df["label"].tolist(),
    bad_words_count=train_df["bad_words_count"].tolist(),
    tokenizer=tokenizer,
    max_length=max_length
)

val_dataset = CyberbullyingDataset(
    texts=val_df["text"].tolist(),
    labels=val_df["label"].tolist(),
    bad_words_count=val_df["bad_words_count"].tolist(),
    tokenizer=tokenizer,
    max_length=max_length
)

test_dataset = CyberbullyingDataset(
    texts=test_df["text"].tolist(),
    labels=test_df["label"].tolist(),
    bad_words_count=test_df["bad_words_count"].tolist(),
    tokenizer=tokenizer,
    max_length=max_length
)

# dataLoaders
train_loader = DataLoader(train_dataset, batch_size=16, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=16, shuffle=False)
test_loader = DataLoader(test_dataset, batch_size=16, shuffle=False)
```

```python
# cyberbullying classifier model: augmented base model with an additional classification layer

class CyberbullyingClassifier(nn.Module):
    def __init__(self, base_model, num_classes):
        super(CyberbullyingClassifier, self).__init__()
        self.base_model = base_model
        self.config = base_model.config
        self.classifier = nn.Sequential(
            nn.Linear(base_model.config.hidden_size + 1, 128),
            nn.ReLU(),
            nn.Dropout(0.2),
            nn.Linear(128, num_classes)
        )

    def forward(self, input_ids=None, attention_mask=None, bad_words_count=None, inputs_embeds=None, labels=None, **kwargs):

        outputs = self.base_model(
            input_ids=input_ids,
            attention_mask=attention_mask,
            inputs_embeds=inputs_embeds,
            **kwargs
        )

        cls_token_embedding = outputs.last_hidden_state[:, 0, :]

        if bad_words_count is not None:
            augmented_features = torch.cat([cls_token_embedding, bad_words_count.unsqueeze(1)], dim=1)
        else:
            augmented_features = cls_token_embedding

        logits = self.classifier(augmented_features)

        if labels is not None:
            loss_fn = torch.nn.CrossEntropyLoss()
            loss = loss_fn(logits, labels)
            return loss, logits

        return logits

#instantiating the augmented model
model = CyberbullyingClassifier(base_model, num_classes=3)
```

```python
# configuring LoRA for self attention layer fine tuning, for sequence classification task

peft_config = LoraConfig(
    task_type=TaskType.SEQ_CLS,
    inference_mode=False,
    target_modules = [
    "self.query",
    "self.key",
    "self.value",
    ],
    r=16,
    lora_alpha=32,
    lora_dropout=0.1
)

# applying the configuration on the model in order to train it
model = get_peft_model(model, peft_config)
```

```python
# setting the optimizer to Adam optimizer and the loss to cross entropy
loss_fn = torch.nn.CrossEntropyLoss()
optimizer = AdamW(model.parameters(), lr=5e-5)
all_true_labels = []
all_preds = []

# allocating the model
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)

# setting the number of epochs
epochs = 10

for epoch in range(epochs):
    print(f"Epoch {epoch + 1}/{epochs}")
    # training the model
    model.train()
    train_loss = 0

    for batch in tqdm(train_loader):
        optimizer.zero_grad()
        input_ids = batch["input_ids"].to(device)
        attention_mask = batch["attention_mask"].to(device)
        labels = batch["label"].to(device)
        bad_words_count = batch["bad_words_count"].to(device)


        logits = model(input_ids=input_ids, attention_mask=attention_mask, bad_words_count=bad_words_count)
        loss = loss_fn(logits, labels)

        # backward pass and optimization
        loss.backward()
        optimizer.step()

        train_loss += loss.item()

    avg_train_loss = train_loss / len(train_loader)
    print(f"Training Loss: {avg_train_loss}")

    # validating the model

    model.eval()
    val_loss = 0
    correct = 0
    total = 0

    with torch.no_grad():
        for batch in val_loader:
            input_ids = batch["input_ids"].to(device)
            attention_mask = batch["attention_mask"].to(device)
            labels = batch["label"].to(device)
            bad_words_count = batch["bad_words_count"].to(device)

            logits = model(input_ids=input_ids, attention_mask=attention_mask, bad_words_count=bad_words_count)
            loss = loss_fn(logits, labels)
            val_loss += loss.item()

            # storing the model predicted values to compute accuracy
            preds = torch.argmax(logits, dim=1)
            correct += (preds == labels).sum().item()
            total += labels.size(0)
            all_true_labels.extend(labels.cpu().numpy())
            all_preds.extend(preds.cpu().numpy())

    avg_val_loss = val_loss / len(val_loader)
    val_accuracy = correct / total
    print(f"Validation Loss: {avg_val_loss}, Validation Accuracy: {val_accuracy}")
    print("\nClassification Report validation:")
    report = classification_report(all_true_labels, all_preds, target_names=["not cyberbullying", "cyberbullying", 'context based'])
    print(report)
```

```python
from sklearn.metrics import classification_report

all_true_labels = []
all_preds = []

print("Testing the model performance on data never seen (test set)")
model.eval()
test_loss = 0
test_correct = 0
test_total = 0
loss_fn = torch.nn.CrossEntropyLoss()
optimizer = AdamW(model.parameters(), lr=5e-5)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)

with torch.no_grad():
    for batch in test_loader:
        input_ids = batch["input_ids"].to(device)
        attention_mask = batch["attention_mask"].to(device)
        labels = batch["label"].to(device)
        bad_words_count = batch["bad_words_count"].to(device)

        logits = model(input_ids=input_ids, attention_mask=attention_mask, bad_words_count=bad_words_count)
        loss = loss_fn(logits, labels)
        test_loss += loss.item()

        preds = torch.argmax(logits, dim=1)
        test_correct += (preds == labels).sum().item()
        test_total += labels.size(0)


        all_true_labels.extend(labels.cpu().numpy())
        all_preds.extend(preds.cpu().numpy())

avg_test_loss = test_loss / len(test_loader)
test_accuracy = test_correct / test_total
print(f"Test Loss: {avg_test_loss}, Test Accuracy: {test_accuracy}")


print("\nClassification Report:")
report = classification_report(all_true_labels, all_preds, target_names=["not cyberbullying", "cyberbullying", 'context based'])
print(report)
```
